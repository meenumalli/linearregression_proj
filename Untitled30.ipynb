```python
def plot_box(df, col_name):
  plt.figure(figsize = (7,7))
  sns.boxplot(df[col_name])
  plt.show()
```

```python
plot_box(df,"Wind speed (m/s)")
```

```python
q1, q2, q3 = np.percentile(df['Wind speed (m/s)'],[25,50,75])
int_q_range = q3 - q1

df = df[df['Wind speed (m/s)'] > (q1 - (int_q_range * 1.5))]
print(q1 - (int_q_range * 1.5))
df = df[df['Wind speed (m/s)'] <(q3 + (int_q_range * 1.5))]
print(q3 + ( int_q_range * 1.5))
# Removing the outliers for the column windspeed
```

```python
df_cor=df.drop(columns=['Rainfall(mm)','Humidity(%)'])
df_cor
# dropping the columns with low correlation
```

```python
from sklearn.model_selection import train_test_split
train, val=train_test_split(df_cor,test_size=0.25)
print(train)


 # splitting the refined data set into train and test data
```

```python
train_x = train.drop(['Rented Bike Count'], axis = 1)
train_y = train['Rented Bike Count']             # output of the training data except the dependent variable

val_x = val.drop(['Rented Bike Count'], axis = 1)# for test data
val_y = val['Rented Bike Count'] # The dependent variable
```

```python
val_y
```

```python
from sklearn.linear_model import LinearRegression

ln_reg = LinearRegression()
x=ln_reg.fit(train_x, train_y)
# The model learns from the training data and predicts the best fit line
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

df_pred = ln_reg.predict(val_x)
print('Mean Absolute Error of linear regression:',mean_absolute_error(df_pred, val_y))
print('Mean Square Error of linear regression:', mean_squared_error(df_pred, val_y))
print('R_Squared Score of linear regression:', r2_score(val_y,df_pred))
# calculating the Metrics
```

```python
submission_data = pd.DataFrame({'Id':val.index,'Pred_bikecount':df_pred,'actual':val_y})
submission_data
```
